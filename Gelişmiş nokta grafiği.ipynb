```python
import matplotlib.pyplot as plt
import numpy as np

# Rastgele veri oluşturma
np.random.seed(0)
x = np.random.rand(100)
y = 2 * x + 1 + np.random.normal(scale=0.2, size=100)  # Doğrusal bir ilişki ekleme
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

# Scatter plot oluşturma
plt.figure(figsize=(12, 8))
scatter = plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis', edgecolor='w', linewidth=0.5)

# Renk skalasını ekleme
plt.colorbar(scatter, label='Renk Skalası')

# Başlık ve etiketler ekleme
plt.title('Gelişmiş Nokta Grafiği', fontsize=15)
plt.xlabel('X Ekseni', fontsize=12)
plt.ylabel('Y Ekseni', fontsize=12)

# Eksen aralıklarını ayarlama
plt.xlim(0, 1)
plt.ylim(0, 3)

# Grid çizgilerini ekleme
plt.grid(True, linestyle='--', alpha=0.7)

# Regresyon çizgisi ekleme
m, b = np.polyfit(x, y, 1)  # Doğrusal regresyon katsayıları
plt.plot(x, m * x + b, color='red', linewidth=2, label=f'y = {m:.2f}x + {b:.2f}')

# Lejant ekleme
plt.legend()

# Veri noktalarına açıklamalar ekleme
for i in range(len(x)):
    plt.text(x[i], y[i], f'{i}', fontsize=9, ha='right', va='bottom')

# Grafik gösterme
plt.show()
```
