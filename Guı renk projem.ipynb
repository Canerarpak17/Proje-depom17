```python
import tkinter as tk
from tkinter import colorchooser

def sec_rengi():
  """Kullanıcının seçtiği rengi alır ve etiketin arka plan rengini değiştirir."""
  renk = colorchooser.askcolor(title="Renk Seç")
  if renk[1]:  # Kullanıcı bir renk seçtiyse
    etiket["bg"] = renk[1]

# Ana pencere oluştur
pencere = tk.Tk()
pencere.title("Renk Seçici")

# Etiket oluştur
etiket = tk.Label(pencere, text="Renk Seç", bg="white", width=20, height=5)
etiket.pack(pady=20)

# Buton oluştur
buton = tk.Button(pencere, text="Renk Seç", command=sec_rengi)
buton.pack()

pencere.mainloop()
```
