```python
import pygame
import sys
import math

# Ekran ve renk ayarları
WIDTH, HEIGHT = 1000, 700
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)

pygame.init()
screen = pygame.display.set_mode((WIDTH, HEIGHT))
clock = pygame.time.Clock()

# Izgara ayarları
cols, rows = 40, 28
spacing_x = int(WIDTH / cols)
spacing_y = int(HEIGHT / rows)
circle_radius = int(min(spacing_x, spacing_y) * 0.9)
dot_radius = 4

# Dönüş parametreleri (dalga kaldırıldı!)
dot_rotation_speed = 0.6

time = 0
running = True

while running:
    screen.fill(WHITE)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    for i in range(cols):
        for j in range(rows):
            x = i * spacing_x + spacing_x // 2
            y = j * spacing_y + spacing_y // 2

            angle = time * dot_rotation_speed + (i + j) * 0.1
            dot_x = x + int(circle_radius * math.cos(angle))
            dot_y = y + int(circle_radius * math.sin(angle))

            pygame.draw.circle(screen, BLACK, (x, y), circle_radius, 2)
            pygame.draw.circle(screen, BLACK, (dot_x, dot_y), dot_radius)

    pygame.display.flip()
    time += 0.05
    clock.tick(60)

pygame.quit()
sys.exit()
```
