```python
import matplotlib.pyplot as plt
import numpy as np

# Veri kümesi oluşturma
data = np.random.randn(2000)

# Histogram oluşturma
plt.hist(data, bins=20, edgecolor='black')

# Grafik başlığı ve eksen etiketleri ekleme
plt.title("Canerin histogram projesi")
plt.xlabel("Değerler")
plt.ylabel("Frekans")

# Histogramı gösterme
plt.show()
```
